```python
#Esercizio 7
#Dal database AdventureWorksDW importiamo le tabelle dimemployee e dimemployeesalesterritory come DataFrame
#Effettuiamo un join tra i due DataFrame usando le colonne EmployeeKey
#Controlliamo la dimensione del DataFrame risultante: è quella attesa?
#Importiamo ora la tabella dimsalesterritory ed effettuiamo un join tra questa e il DataFrame risultante della join precedente, usando le colonne SalesTerritoryKey
#Su questo DataFrame contare quanti dipendenti ci sono per ogni paese (country) e per ogni regione (region)
#Valutiamo la media del BaseRate per ogni paese: ci sono differenze?
```

```python
import os
from dotenv import load_dotenv
import sqlalchemy
import pandas as pd

env_path = "C:\\Users\\Franco\\cred.env" 
load_dotenv(dotenv_path=env_path, override=True)

username = os.getenv("username")
password = os.getenv("password")
host = os.getenv("host")
dbname = os.getenv("dbname")

connection_string = f"mysql+pymysql://{username}:{password}@{host}/{dbname}"

db_engine = sqlalchemy.create_engine(connection_string)
```

```python
query_employee = "SELECT * FROM dimemployee"
df_employee = pd.read_sql(query_employee, db_engine)
df_employee
```

```python
query_employeesalesterritory = "SELECT * FROM dimemployeesalesterritory"
df_employeesalesterritory = pd.read_sql(query_employeesalesterritory, db_engine)
df_employeesalesterritory
```

```python
#Metodo MERGE
df_employee_merge = df_employee.merge(df_employeesalesterritory, how = "inner", on = "EmployeeKey")
df_employee_merge
```

```python
query_salesterritory = "SELECT * FROM dimsalesterritory"
df_salesterritory = pd.read_sql(query_salesterritory, db_engine)
df_salesterritory
```

```python
#Metodo MERGE
df_merge = df_employee_merge.merge(df_salesterritory, how = "inner", on = "SalesTerritoryKey")
df_merge
```

```python
#Conto il numero di dipendenti per ogni country
dipendenti_country = df_merge.groupby("SalesTerritoryCountry")["EmployeeKey"].value_counts()
dipendenti_country
```

```python
#Conto il numero di dipendenti per ogni regione
dipendenti_region = df_merge.groupby("SalesTerritoryRegion")["EmployeeKey"].value_counts()
dipendenti_region
```

```python
#Calcolo la media totale del baserate
media_baserate_totale = df_merge["BaseRate"].mean()
media_baserate_totale
```

```python
#Valuto la media del BaseRate per ogni paese
media_baserate = df_merge.groupby("SalesTerritoryCountry")["BaseRate"].mean()
media_baserate #la media più alta si presenta per i paesi di Australia, United Kingdom e Germany, con uno stesso valore. La più bassa si presenta per United States, 
#mentre Canada e France si presentano al secondo posto con uno stesso valore. Questi paesi sono quelli che si avvicinano di più al valore della media totale
```
