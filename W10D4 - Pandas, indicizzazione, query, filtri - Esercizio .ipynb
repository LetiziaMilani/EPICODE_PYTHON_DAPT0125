Esercizio 1.3
Qual è il prezzo medio per i prodotti di colore blu?
Qual è il prezzo medio per i prodotti di colore rosso o nero?
Qual è il prezzo massimo per i prodotti di taglia 42 e peso oltre i 10 Kg?
Qual è il nome inglese e il costo di produzione (StandardCost) di tutti i prodotti di taglia 42, peso oltre i 10 Kg e colore argento?
Esaminiamo la differenza tra lo StandardCost e il DealerPrice degli ultimi 20 elementi del dataset: c'è un pattern? C'è qualche elemento che non lo segue?

```python
#prezzo medio prodotti di colore blu
filtro_colore = df["Color"] == "Blue"
subset_colore_blu = df.loc[filtro_colore]
prezzo_medio_blu = subset_colore_blu["DealerPrice"].mean()
prezzo_medio_blu
```

```python
#prezzo medio prodotti colore rosso o nero
filtro_colore_rosso = df["Color"] == "Red"
filtro_colore_nero = df["Color"] == "Black"
filtro_colore_rosso_nero = filtro_colore_rosso | filtro_colore_nero
subset_colore_rosso_nero = df.loc[filtro_colore_rosso_nero]
prezzo_medio_rosso_nero = subset_colore_rosso_nero["DealerPrice"].mean()
prezzo_medio_rosso_nero
```

```python
#prezzo massimo per i prodotti di taglia 42 e peso oltre i 10 Kg
filtro_taglia = df["Size"] == "42"
filtro_peso = df["Weight"] > 10
filtro_taglia_peso = filtro_taglia & filtro_peso
subset_taglia_peso = df.loc[filtro_taglia_peso]
prezzo_massimo = subset_taglia_peso["DealerPrice"].max()
prezzo_massimo
```

```python
#nome inglese e il costo di produzione di tutti i prodotti di taglia 42, peso oltre i 10 Kg e colore argento
filtro_colore_argento = df["Color"] == "Silver"
filtro_taglia_peso_colore = filtro_taglia & filtro_peso & filtro_colore_argento
subset_taglia_peso_colore = df.loc[filtro_taglia_peso_colore, ["ProductKey", "EnglishProductName", "StandardCost"]]
subset_taglia_peso_colore
```

```python
#differenza tra lo StandardCost e il DealerPrice degli ultimi 20 elementi del dataset
ultimi_venti_elementi_dataset = df.tail(20)
subset_ultimi_venti_elementi = ultimi_venti_elementi_dataset[["ProductKey", "StandardCost", "DealerPrice"]]

ultimi_venti_elementi_dataset["Differenza"] = ultimi_venti_elementi_dataset["StandardCost"] - ultimi_venti_elementi_dataset["DealerPrice"]

subset_ultimi_venti_elementi = ultimi_venti_elementi_dataset[["ProductKey", "StandardCost", "DealerPrice", "Differenza"]]
subset_ultimi_venti_elementi 

#tutti i prodotti presentano una differenza negativa ad eccezione per gli ultimi tre per la quale risulta positiva
```

```python
#dati relativi a diverse fasce di età: <20, 20-30, 30-40, 40-50, >50;

filtro_eta_minore_20 = df["Age (years)"] < 20
subset_eta_minore_20 = df.loc[filtro_eta_minore_20]
subset_eta_minore_20

filtro_eta_tra_20_30 = df["Age (years)"].between(20, 30)
subset_eta_tra_20_30 = df.loc[filtro_eta_tra_20_30]
subset_eta_tra_20_30

filtro_eta_tra_30_40 = df["Age (years)"].between(30, 40)
subset_eta_tra_30_40 = df.loc[filtro_eta_tra_30_40]
subset_eta_tra_30_40

filtro_eta_tra_40_50 = df["Age (years)"].between(40, 50)
subset_eta_tra_40_50 = df.loc[filtro_eta_tra_40_50]
subset_eta_tra_40_50

filtro_eta_maggiore_50 = df["Age (years)"] > 50
subset_eta_maggiore_50 = df.loc[filtro_eta_maggiore_50]
subset_eta_maggiore_50
```

```python
#media pressione diastolica per fasce età
media_pressione_eta_tra_20_30 = subset_eta_tra_20_30["Diastolic blood pressure (mm Hg)"].mean()
media_pressione_eta_tra_20_30

media_pressione_eta_tra_30_40 = subset_eta_tra_30_40["Diastolic blood pressure (mm Hg)"].mean()
media_pressione_eta_tra_30_40

media_pressione_eta_tra_40_50 = subset_eta_tra_40_50["Diastolic blood pressure (mm Hg)"].mean()
media_pressione_eta_tra_40_50

media_pressione_eta_maggiore_50 = subset_eta_maggiore_50["Diastolic blood pressure (mm Hg)"].mean()
media_pressione_eta_maggiore_50
```

```python
#media della pressione per ogni singolo anno di età
media_pressione_per_eta = df.groupby("Age (years)", as_index=False)["Diastolic blood pressure (mm Hg)"].mean()
media_pressione_per_eta
```
