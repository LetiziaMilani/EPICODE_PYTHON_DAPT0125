Esercizio 2
Creiamo un DataFrame sintetico, che contiene i guadagni mensili di diverse annate, con il seguente codice: 
years = 5 
guadagni = pd.DataFrame({"Mese": list("GFMAMGLASOND"*years), "Anno": np.repeat(list(range(years)), 12), "Valore": np.random.randint(800, 5000, 12*years)})
Calcola la somma cumulativa dei guadagni utilizzando il metodo .cumsum()
Come sopra, ma raggruppato per ogni anno usando prima un .groupby()

```python
import numpy as np
import pandas as pd

years = 5
guadagni = pd.DataFrame({"Mese": list("GFMAMGLASOND"*years), "Anno": np.repeat(list(range(years)), 12), "Valore": np.random.randint(800, 5000, 12*years)})
guadagni
```

```python
#Calcolo somma cumulativa dei guadagni
somma = guadagni["Valore"].cumsum()
somma 
```

```python
#Calcolo somma cumulativa dei guadagni per anno
somma_per_anno = guadagni.groupby("Anno")["Valore"].cumsum()
somma_per_anno
```

Quali righe hanno un valore nullo nella colonna num-of-doors?
Esaminando i dati nel dataset, cerchiamo una logica per sostituire i valori nulli nella colonna num-of-doors

```python
import pandas as pd
df = pd.read_csv("C:\\Users\\Franco\\Desktop\\Leti\\automobile.csv")
df
```

```python
#Verifico se ci sono valori nulli
df.isnull()
```

```python
#Verifico in quali colonne sono presenti i valori nulli
df.isnull().sum() #sono presenti 37 valori nulli nella colonna normalized-losses e 2 nella colonna num-of-doors
```

```python
filtro = df["num-of-doors"].isnull()
righe_con_null = df.loc[filtro]
righe_con_null #Verifico quali sono le righe che presentano valori nulli per la colonna num-of-doors
```

```python
#Verifico se le per i valori di symboling pari a 1 o a 0 il valore di num-of-doors è sempre lo stesso
filtro_zero = df["symboling"] == 0
df_filtro_zero = df.loc[filtro_zero, "num-of-doors"]
df_filtro_zero 
```

```python
filtro_uno = df["symboling"] == 1
df_filtro_uno = df.loc[filtro_uno, "num-of-doors"]
df_filtro_uno
```

```python
#Dato che per i valori di symboling pari a 0 e a 1 per cui il valore assunto da num-of-doors, verifico quale è il valore per num-of-doors che compare più volte
conteggio = df.groupby(["symboling", "num-of-doors"]).size()
conteggio #per i valori di symboling uguale a 0, il valore che compare maggiormente per num-of-doors è four, mentre per valori pari a 1 il valore che compare maggiormente per num-of-doors è two 
```

```python
#Vado a sostituire, per la riga in cui symboling uguale a 0, num-of-doors = four, per la riga in cui symboling uguale a 1, num-of-doors = two
df.loc[filtro_zero, "num-of-doors"] = df.loc[filtro_zero, "num-of-doors"].fillna("four")
df.loc[filtro_uno, "num-of-doors"] = df.loc[filtro_uno, "num-of-doors"].fillna("two")
```

```python
df.isnull().sum() #Verifico che nella colonna num-of-doors non ci siano più valori nulli 
```

Esercizio 8
Nel pacchetto os della standard library c'è la funzione os.listdir() che permette di avere la lista dei nomi di file all'interno di una directory; 
senza input di default li cerca nella directory di lavoro corrente, altrimenti si può passare un path per esaminare una directory specifica, 
ad esempio os.listdir("mio_progetto/beginner_datasets/")
Nella directory dei beginner_datasets, quali sono i dataset che contengono dati nulli?

```python
import os
import pandas as pd
```

```python
# Lista per memorizzare i nomi dei file con dati nulli
file_con_nulli = []

# Cartella da esplorare
cartella = "C:\\Users\\Franco\\Desktop\\Leti\\beginner_datasets\\"

# Ciclo su tutti i file nella cartella
for nome_file in os.listdir(cartella):
    
    # Se il file è un .csv
    if nome_file.endswith(".csv"):
        
        # Costruisce il path completo al file
        percorso_completo = os.path.join(cartella, nome_file)
        
        # Legge il file CSV
        df = pd.read_csv(percorso_completo)
        
        # Conta i valori nulli nel DataFrame
        numero_nulli = df.isna().sum().sum()  # somma su righe e colonne
        
        # Se ci sono valori nulli, aggiunge il nome alla lista
        if numero_nulli > 0:
            file_con_nulli.append(nome_file)

# Stampa i risultati
print("File con dati nulli:")
print(file_con_nulli)
```
